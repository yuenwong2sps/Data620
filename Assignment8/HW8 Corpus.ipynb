## Choose a corpus of interest.

#corpora is a large bodies of text data. Many corpora are designed to contain a careful balance of material in one or more genres. 

```python
import nltk
```

```python
import pandas as pd
import numpy as py
```

```python
nltk.download('gutenberg')
```

```python
# file identifiers in this corpus
files=nltk.corpus.gutenberg.fileids()
```

```python
files
```

```python
words=[]
for f in files:
    tokens =nltk.corpus.gutenberg.words(f)
    for w in tokens:
        if w.isalpha():
            words.append(w)
```

```python
len(words)
```

```python
worddf=pd.DataFrame(words, columns=['word'])
unique_w=worddf.word.unique()
```

### How many total unique words are in the corpus? (Please feel free to define unique words in any interesting, defensible way).

```python
len(unique_w)
```

### Taking the most common words, how many unique words represent half of the total words in the corpus?Identify the 200 highest frequency words in this corpus.

```python
from collections import Counter
```

```python
keys=Counter(worddf.word)
keys
```

```python
#sort from top hightest counts
sorted_keys = sorted(keys, key=keys.get, reverse=True)
```

```python
word_list=[]
for k in sorted_keys[:200]:
    word_list.append(k)
```

```python
print(word_list)
```

```python
fdist=nltk.FreqDist(words)
```

```python
word_list_count = []
for m in word_list:
    word_list_count.append(fdist[m])
    print( m + ':' , fdist[m])
```

### Does the observed relative frequency of these words follow Zipf’s law? Explain.

According to the Zipf's law, the count of the first word is double of the count of the second word, and three times as often as the third word. 

It is true in our example, the first word "the" is nearly double of the second word "and", however, it is not three times as often as the third word "of"

### In what ways do you think the frequency of the words in this corpus differ from “all words in all corpora.”


```python
c = list(range(201))
c = c[1:201]
freq_df = pd.DataFrame(data=c,index=word_list_count)
```

### Rank vs frequency

```python

import matplotlib.pyplot as plt
plt.plot(c, word_list_count, 'ro')
plt.show()

```

```python
freq_df.plot(loglog=True)
plt.show()
```

From the chart above the log of freqency and log of word rank has nearly linear relationship.
